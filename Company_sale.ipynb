```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
path = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv'
```

```python
data = pd.read_csv(path, sep = ',')
```

```python
plt.plot(data.month_number, data.total_profit)

plt.xlabel('Month number')
plt.ylabel('Profit')

plt.xticks(data.month_number.tolist())
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title('Company profit per month');
```

```python
plt.plot(data.month_number, 
         data.total_profit, ls = '--', 
         lw= 3, c= 'r', marker = 'o', markerfacecolor = 'k',
         label = 'Profit data of last year')

plt.xlabel('Month number')
plt.ylabel('Profit')

plt.xticks(data.month_number.tolist())
plt.yticks([100000, 200000, 300000, 400000, 500000])

plt.legend(loc = 'lower right')
plt.title('Company Sales data of last year');
```

```python
plt.plot(data.month_number, data.facecream, c = 'blue', marker = 'o', lw = 3, label = 'Face cream sales data')
plt.plot(data.month_number, data.facewash, c = 'orange', marker = 'o', lw = 3, label = 'Face wash sales data')
plt.plot(data.month_number, data.toothpaste, c = 'green', marker = 'o', lw = 3, label = 'Tooth paste sales data')
plt.plot(data.month_number, data.bathingsoap, c = 'r', marker = 'o', lw = 3, label = 'Bathingsoap sales data')
plt.plot(data.month_number, data.shampoo, c = 'purple', marker = 'o', lw = 3, label = 'Shampoo sales data')
plt.plot(data.month_number, data.moisturizer, c = 'brown', marker = 'o', lw = 3, label = 'moisturizer sales data')
 
plt.xticks(data.month_number.tolist())
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])

plt.title('Sales data')
plt.ylabel("Sales unit in number")
plt.xlabel('Month number')

plt.legend(loc = 'upper left');
```

```python
plt.scatter(data.month_number, data.toothpaste, label = 'Tooth paste Sales data')
plt.xticks(data.month_number.tolist())
plt.grid(ls = '--')
plt.legend();
```

```python
month = data.month_number.tolist()
```

```python
plt.bar(month, data.facecream, width = -0.3, align = 'edge', label = 'Face crean sales data')
plt.bar(month, data.facewash, width = 0.3, align = 'edge', label = 'Face wash sales data')

plt.xticks(month)
plt.legend(loc = 'upper left')
plt.grid(lw = 1, ls = '--')

plt.title("Facewash and facecream sales data")
plt.ylabel('Sales unit in number')
plt.xlabel('Month Number');
```

```python
plt.bar(month, data.bathingsoap)
plt.xticks(month)
plt.ylabel('Sales unit in number')
plt.xlabel('Month Number');
plt.grid(ls = '--', lw = 1)
plt.title("bathingsoap sales data")
```

```python
range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]

plt.hist(data.total_profit, range, label = 'Profit data')
plt.xlabel("Aactual Profit in dollar")
plt.ylabel('profit range in dollar')
plt.xticks(range)
plt.title('Profit data')
plt.legend(loc = 'upper left');
```

```python
plt.figure(figsize = (10, 6))
facecream = data.facecream.sum()
facewash = data.facewash.sum()
toothpaste = data.toothpaste.sum()
bathingsoap = data.bathingsoap.sum()
shampoo = data.shampoo.sum()
moisturizer = data.moisturizer.sum()

sales = [facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
labels = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

plt.pie(sales, labels = labels, autopct='%1.1f%%')
plt.legend(loc = 'lower right')
```

```python
plt.figure(figsize = (10, 6))
fig, ax = plt.subplots(2, sharex = True)
ax[0].plot(month, data.bathingsoap, marker = 'o', c = 'b', lw = 3)
ax[0].set_title('Sales data of a Bathingsoap')

ax[1].plot(month, data.facewash, marker = 'o', c = 'r', lw = 3)
ax[1].set_title('Sales data of face wash')
```

```python
plt.stackplot(month, data.facecream, data.facewash, data.toothpaste, data.bathingsoap, data.shampoo, data.moisturizer, colors=['blue', 'orange', 'pink', 'green', 'purple', 'c'])
```
